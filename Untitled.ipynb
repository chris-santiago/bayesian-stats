# Models

## Base Linear Model

```r
mod.base = lm(weight ~ feed_name + 0, data=data)
summary(mod.base)
```

## Pooled Model

```r
mod_string = " model {
  for (i in 1:length(weight)) {
    weight[i] ~ dnorm(mu, sigma)
  }
  
  mu ~ dgamma(10, 1/10)
  sigma ~ dlnorm(1, 1)
} "

set.seed(43)
data_jags = as.list(data)
params = c("mu", "sigma")

mod.pooled = run_sim(
  model.str = mod_string,
  data = data_jags,
  params = params
)
```

```r
(dic.pooled = dic.samples(mod.pooled$obj, n.iter=1e3))
```

```r
round(mod.pooled$coefs, 4)
```

```r
plot(mod.pooled$csim)
```

## Hierarchical Model

```r
mod_string = " model {
  for (i in 1:length(weight)) {
    weight[i] ~ dnorm(theta[feed[i]], sigma)
  }
  
  for (j in 1:max(feed)) {
    theta[j] ~ dnorm(mu, tau)
  }
  
  mu ~ dgamma(10, 1)
  tau ~ dlnorm(1, 1)
  sigma ~ dlnorm(1, 1)
  
} "

set.seed(43)
data_jags = as.list(data)
params = c("theta", "mu", "sigma")

mod.hier = run_sim(
  model.str = mod_string,
  data = data_jags,
  params = params
)
```

```r
(dic.hier = dic.samples(mod.hier$obj, n.iter=1e3))
```

```r
round(mod.hier$coefs, 4)
```

```r
plot(mod.hier$csim, ask=FALSE)
```

## Heterogenous Model

```r
mod_string = " model {
  for (i in 1:length(weight)) {
    weight[i] ~ dnorm(theta[feed[i]], sigma)
  }
  
  for (j in 1:max(feed)) {
    theta[j] ~ dnorm(mu[j], tau)
  }

  for (j in 1:max(feed)) {
    mu[j] ~ dnorm(phi, 1e6)
  }
  
  phi ~ dexp(1/20)
  tau ~ dlnorm(1, 1)
  sigma ~ dlnorm(1, 1)
  
} "

set.seed(43)
data_jags = as.list(data)
params = c("theta", "mu", "sigma")

mod.hetero = run_sim(
  model.str = mod_string,
  data = data_jags,
  params = params
)
```

```r
(dic.hetero = dic.samples(mod.hetero$obj, n.iter=1e3))
```

```r
round(mod.hetero$coefs, 4)
```
